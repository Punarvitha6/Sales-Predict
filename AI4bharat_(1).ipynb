```python
def get_max_sequence_length(df, english_column='English', hindi_column='Hindi'):

  # Find maximum length and sentence with maximum length in English column
  max_length_english = df[english_column].str.split().str.len().max()
  max_index_english = df[english_column].str.split().str.len().idxmax()
  max_sentence_english = df.loc[max_index_english, english_column]

  # Find maximum length and sentence with maximum length in Hindi column
  max_length_hindi = df[hindi_column].str.split().str.len().max()
  max_index_hindi = df[hindi_column].str.split().str.len().idxmax()
  max_sentence_hindi = df.loc[max_index_hindi, hindi_column]

  # Return the results for English and Hindi
  return max_length_english, max_sentence_english, max_length_hindi, max_sentence_hindi

```

```python
# Find the maximum sequence lengths and corresponding sentences
max_length_english, max_sentence_english, max_length_hindi, max_sentence_hindi = get_max_sequence_length(data)

print(f"Maximum sequence length (English): {max_length_english}")
print(f"Sentence with maximum length (English): {max_sentence_english}")
print(f"\nMaximum sequence length (Hindi): {max_length_hindi}")
print(f"Sentence with maximum length (Hindi): {max_sentence_hindi}")
```

```python
import pandas as pd


def filter_by_sequence_length(dataframe, english_column='English', hindi_column='Hindi', threshold=50):

  # Get original sequence lengths (improved readability)
  english_lengths = dataframe[english_column].str.split().str.len()
  hindi_lengths = dataframe[hindi_column].str.split().str.len()

  # Apply vectorized string length checks with logical OR for filtering
  filtered_df = dataframe[(english_lengths <= threshold) & (hindi_lengths <= threshold)]

  return filtered_df, english_lengths, hindi_lengths

```

```python
# Example usage
filtered_df, english_lengths, hindi_lengths = filter_by_sequence_length(data.copy())

print(f"Shape of the original DataFrame: {data.shape}")
print(f"Shape of the filtered DataFrame: {filtered_df.shape}")
```

```python
# Find the maximum sequence lengths and corresponding sentences
max_length_english, max_sentence_english, max_length_hindi, max_sentence_hindi = get_max_sequence_length(filtered_df)

print(f"Maximum sequence length (English): {max_length_english}")
print(f"Sentence with maximum length (English): {max_sentence_english}")
print(f"\nMaximum sequence length (Hindi): {max_length_hindi}")
print(f"Sentence with maximum length (Hindi): {max_sentence_hindi}")
```

```python
from transformers import MBartForConditionalGeneration, MBartTokenizer, TrainingArguments, Trainer
from transformers import AlbertTokenizer, AutoTokenizer
from datasets import Dataset

# Load your cleaned English and Hindi data (replace with your data source)
english_data = data["English"].to_list()
hindi_data = data["Hindi"].to_list()

# Load the tokenizer and model
model_name = "ai4bharat/IndicBART"
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/IndicBART")

model = MBartForConditionalGeneration.from_pretrained(model_name)


```

```python
# Tokenize the data (adjust batch size based on memory constraints)
max_length = 52  # Adjust this value as needed

en_tokenized = tokenizer(
    english_data,
    padding="max_length",
    truncation=True,
    max_length=max_length,
    return_tensors="pt",
)

hi_tokenized = tokenizer(
    hindi_data,
    padding="max_length",
    truncation=True,
    max_length=max_length,
    return_tensors="pt",
)


# Combine data
data = {
    "input_ids": en_tokenized["input_ids"],
    "attention_mask": en_tokenized["attention_mask"],
    "labels": hi_tokenized["input_ids"],  # Hindi IDs as labels (teacher forcing)
}

```

```python
num_datapoints = len(data['input_ids'])  # Assuming all keys have the same length
split_point = int(0.8 * num_datapoints)  # Split point for 80% training data
```

```python
train_data = {}
validation_data = {}

# Initialize each key with an empty tensor
for key in data.keys():
  train_data[key] = torch.zeros_like(data[key][:split_point])
  validation_data[key] = torch.zeros_like(data[key][split_point:])
```

```python
for key, value in data.items():
  train_data[key] = value[:split_point]
  validation_data[key] = value[split_point:]
```

```python
len(train_data['input_ids'])
```

```python
# Create datasets
train_dataset = Dataset.from_dict(train_data)
validation_dataset = Dataset.from_dict(validation_data)
```

```python
#pip install accelerate -U
```

```python
# Split data into training and validation sets (assuming you have a split point)
batch_size=8

# Define training arguments (adjust hyperparameters as needed)
training_args = TrainingArguments(
    output_dir="./outputs",
    per_device_train_batch_size=batch_size,  # Adjust based on GPU memory
    save_steps=5000,
    num_train_epochs=1,  # Adjust epochs
    learning_rate=2e-5,
)

# Create the Trainer object
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=validation_dataset,
)

```

```python
# Train the model
trainer.train()

```

```python
# Save the fine-tuned model
trainer.save_model("./fine-tuned_indicbart_en_hi")
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Assuming you saved your trained model directory (replace with your path)
model_name = "/content/drive/MyDrive/ai4bharat/fine-tuned_indicbart_en_hi"  # Update with the actual directory

# Load the IndicBART tokenizer (ensure compatibility)
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/IndicBART")

# Load the fine-tuned model (use AutoModelForSeq2SeqLM for IndicBART)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)


```

```python
# Example source text (replace with your actual text)
source_text = "Hi How are you"

# Tokenize the source text using the IndicBART tokenizer
encoded_input = tokenizer(source_text, return_tensors="pt")
input_ids = encoded_input["input_ids"]  # Access tokenized input IDs

# Generate translation with beam search
with torch.no_grad():
  generated_ids = model.generate(
      input_ids=input_ids,
      max_length=50,  # Adjust maximum output length
      num_beams=5  # Adjust number of beams for beam search
  )

# Decode the generated tokens
translated_text = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(f"Translated Text: {translated_text}")
```
